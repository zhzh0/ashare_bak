```python
def get_ratio(p):  # p传入close列， 返回标的涨跌幅
    result = []
    pre = -1
    for i in p:
        if pre == -1:
            pre = i
            continue
        else:
            result.append((i/pre-1)*100)
            pre = i
    return result
```

```python
cols = ['sh000300','sh000905','sz399006','sh513050','sh513500','sh513100','sh510050']
#cols = ['sh513500','sh513100']
df = pd.DataFrame()
for c in cols:
    tmp = get_price(c,frequency='1d',count=200) 
    val = get_ratio(tmp['close'])
    df[c] = val

sns.heatmap(df.corr(),annot=True)
```
